```python
#import requests and beautiful soup in order to scrape
import requests
from bs4 import BeautifulSoup
```

```python
#generating list of universities to then use to scrape other websites
page = requests.get("https://www.4icu.org/us/a-z/")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
#soup_1 gets the areas of html where university name is stored
soup_1 = soup.find('tbody').find_all('a')
```

```python
soup_1
```

```python
#getting university name from each line of html and adding it to list
list_of_colleges = [line.get_text() for line in soup_1]
```

```python
list_of_colleges
```

```python
#function for formatting list of colleges in order to call in independent url
def format_col_for_call(list_):
    cleaned_list = [col.replace('&','').replace(' ','-') for col in list_]
    return cleaned_list
```

```python
formatted_list_of_col = format_col_for_call(list_of_colleges)
```
